### Numpy files for Deep Learning

- Converts the images into numpy arrays.

```python
import numpy as np
import pandas as pd
from astropy.table import Table
from PIL import Image
import os
import seaborn as sns; sns.set()
sns.set_theme(style="ticks")
```

```python
# Directory containing your images
images_directory = '../legacy_color_images'  # Replace with your actual directory
```

```python
# Create a folder to store the numpy arrays if it doesn't exist
output_directory = '../Data_training'  
os.makedirs(output_directory, exist_ok=True)
# Example: Combining images for class 1
```

```python
# Get a list of all image files in the directory
image_files = [file for file in os.listdir(images_directory) if file.endswith(('jpg', 'jpeg', 'png'))]
```

```python
image_files
```

```python
# Initialize an empty list to store images for class 1
class_1_images = []
```

```python
# Loop through each image file for class 1
for filename in image_files:
    image_path = os.path.join(images_directory, filename)
    
    # Open and preprocess the image
    img = Image.open(image_path)
    img = img.resize((64, 64))  # Resize the image to 64x64
    img = np.array(img) / 255.0  # Normalize pixel values to [0, 1]
    
    # Append the preprocessed image to the list for class 1
    class_1_images.append(img)
```

```python
# Convert the list of images for class 1 to a numpy array
class_1_data = np.array(class_1_images)

# Save the combined array for class 1
np.save(os.path.join(output_directory, 'class_1_combined.npy'), class_1_data)
```
